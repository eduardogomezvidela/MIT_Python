Branching program:

If, elif, else



### While loop

Used when we don't know when we will break out of the loop

```python
x=0
while x < 10:
    print(x)
    x+=1
```

### For loop

Used when we know when we will break out of the loop

```python
for i in range(10):#prints 0 to 9 inclusive
    print(i)
```

```python
for i in range(0,4):#prints 0 to 3 inclusive
    print(i)
```

```python
for i in range(0,10,2):#skips some of the values
    print(i)
```

```python
my_sum = 0
for i in range(5):
    my_sum += i
print(my_sum)
```

### Break

Will break out of the for loop (terminate it)

```python
for i in range(0,10):
    print(i)
    if i == 6:
        break
```
